```python
dic = {"Year": [], "Price":[]}
for tr in soup.find_all('tr')[4:4+2011-1949+1]:
    tds = tr.find_all('td')
    year = tds[0].text
    if year < '1990':
        val = tds[1].text.strip('\r\n')
        aval = val.strip('R ')
    else:
        val = tds[3].text.strip('\r\n')
        aval = val.strip('R')
    #dic = {"Year":[int(key)], "Price": [float(aval)]}
    dic["Year"].append(int(year))
    dic["Price"].append(float(aval))
    
   

        
    
```
